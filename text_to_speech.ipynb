```python
# Make sure Coqui TTS is installed and upgraded
from TTS.api import TTS # Import the TTS class from the Coqui TTS library

# Download and use Tacotron2 model with a vocoder
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", # Instantiate the TTS class
          gpu=False)  # Assuming you are using CPU; change to True if using GPU

# The vocoder is automatically selected based on the model.
# If you want to specify a vocoder explicitly, you might need to
# use a different method or a more recent version of the TTS library.
tts.tts_to_file(text="Hello, this is Tacotron 2 in action.", # Call tts_to_file on the TTS instance
                file_path="tacotron_output.wav")

print("Tacotron audio saved as tacotron_output.wav")
```
